Model :- 1

The relation between price and size

```python
import matplotlib.pyplot as plt
```

```python
# converting the data set int test data and train data
# train is 80% and test is 20%
from sklearn.model_selection import train_test_split
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(df[['SIZE']], df['PRICE'], train_size = 0.8)

#getting shape of the train data and test data
print("Training data X_train for Size and Price : ",X_train_1.shape)
print("Training data y_train for Size and Price : ",y_train_1.shape)
print("Testing data X_test for Size and Price : ",X_test_1.shape)
print("Testing data X_test for Size and Price : ",y_test_1.shape)
```

```python
from sklearn import model_selection

# importing linear regression,r2_score,mean_squared_error from sklearn
# importing math

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import math


# linear model for train data
model_lin_1=LinearRegression().fit(X_train_1,y_train_1)

# Prediction for the test data
pred_lin_1 = model_lin_1.predict(X_test_1)

# The scatter plot with test data
plt.scatter(X_test_1, y_test_1, color="red")
# getting plot
plt.plot(X_test_1, pred_lin_1, color="green", linewidth=3)

# X label as X_test
plt.xlabel("X_test")
# Y label as y_test
plt.ylabel("y_test")

plt.xticks(())
plt.yticks(())

plt.show()


Root_Mean_Sq_Er=math.sqrt(mean_squared_error(y_test_1,pred_lin_1))

# printing the regression score for the train data
print("The coeff. of determined prediction value for Size and Price  :-",model_lin_1.score(X_train_1,y_train_1))
# The coefficients
print("Coeff. for Size and Price : ", model_lin_1.coef_)
# The mean squared error
print("Mean squared error: %.2f for Size and Price" % mean_squared_error(y_test_1,pred_lin_1))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination for Size and Price : %.2f" % r2_score(y_test_1,pred_lin_1))
# root mean square error
print("Root Mean Square Error for Size and Price :- ",Root_Mean_Sq_Er)
```

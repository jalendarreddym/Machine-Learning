Principal component analysis clustering

```
from sklearn.decomposition import PCA
```

```
P_C_A = PCA(n_components=1)
X_train = P_C_A.fit_transform(df_train)
```

```
from sklearn.preprocessing import StandardScaler
```

```
stdrd_scaler = StandardScaler()
stdrd_scaler.fit_transform(X_train)
```

```
from sklearn.cluster import KMeans
```

```
K_M_models = [KMeans(n_clusters=k, random_state=25).fit(X_train) for k in range (1, 15)]
K_M_inertia = [model.inertia_ for model in K_M_models]

figure, a_x = plyt.subplots(figsize=(25,15))
plyt.plot(range(1, 15), K_M_inertia)
plyt.title('Elbow method for KMEANS')
plyt.xlabel('No. of Clusters')
plyt.ylabel('Within-Cluster Sum of Square')
plyt.show()
```

```
print("The maximum accuracy occurs from the K value is 2(K=2) to the K value is 4(K=4) from the graph above")
```

```
def K_M_model(X, n_neighbors):
    K_M_model = KMeans(n_neighbors)
    K_M_model.fit(X)
    K_M_clust_labels = K_M_model.predict(X)
    K_M_center = K_M_model.cluster_centers_
    return (K_M_clust_labels, K_M_center)
```

```
# K=2
K=2
K_M_clust_label_1, K_M_center_1 = K_M_model(X_train, K)
K_M_1 = pd.DataFrame(K_M_clust_label_1)
```

```
from sklearn.metrics import silhouette_score
```

```

# For k=2

K_M_S_2= KMeans(n_clusters=2, random_state=25)
K_M_S_2.fit(X_train)

print('The silhouette coefﬁcient value of our model is ' + str(round(silhouette_score(X_train, K_M_S_2.labels_),2)))
```
